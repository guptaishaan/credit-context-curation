# Credit-risk temporal EDA

Run this notebook from the project root after placing the Kaggle files in `data/`. It checks date coverage and annual default rates before the GPU experiments.

```python
import sys
from pathlib import Path
import matplotlib.pyplot as plt
sys.path.insert(0, str(Path.cwd() / 'src'))
from data_loader import load as load_lending_club
from data_loader_hc import load as load_home_credit
```

```python
for name, loader, path in [('Lending Club', load_lending_club, 'data/lending_club.csv'), ('Home Credit', load_home_credit, 'data/home_credit_base.csv')]:
    X, y, dates = loader(path)
    annual = y.groupby(dates.dt.year).agg(['size', 'mean'])
    print(f'\n{name}\n', annual)
    annual['mean'].plot(marker='o', label=name)
plt.title('Annual default rate'); plt.xlabel('Year'); plt.ylabel('Default rate'); plt.legend(); plt.show()
```
